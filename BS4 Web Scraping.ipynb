```python
rows = soup.find_all('tr')
rows[:10]
```

```python
# get all table rows in list form first and then convert that list into a dataframe
for row in rows:
    row_td = row.find_all('td')
print(row_td)
print(type(row_td))
```

```python
# use the get_text() method to extract the text without html tags
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
# view column '0'
print(df[0])
```

```python
# DATA MANIPULATION AND CLEANING
# input data into a new column after every ','
df1 = df[0].str.split(',', expand=True)
df1.head(10)
```

```python
# remove square brackets
df1[0] = df1[0].str.strip('[]')
df1.head(10)
```

```python
# add missing table headers
col_labels = soup.find_all('th')
print(col_labels)
```

```python
# do not want the tags, remove using get_text() after converting to str first then parsing
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
print(cleantext2)
```

```python
headers = []
headers.append(cleantext2)
headers
```

```python
# convert headers into a df
df2 = pd.DataFrame(headers)
df2.head()

```

```python
# but we want a separate column for each header
df3 = df2[0].str.split(',', expand=True)
df3
```

```python
# integrate the headers into original df
df4 = pd.concat([df3, df1])
df4.head(10)
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head(30)
```

```python
time_list = df7[' Chip Time'].tolist()
# print(time_list)

# You can use a for loop to convert 'Chip Time' to minutes
time_mins = []
for i in time_list:
    h, m, s = i.split(':')
    math = (int(h) * 3600 + int(m) * 60 + int(s))/60
    time_mins.append(math)
# print(time_mins)
```
